```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
pi=np.pi
x=np.linspace(-4*pi,4*pi,1000)
plt.plot(x,np.sin(x)/x)
plt.show()
```

```python
x=np.linspace(0,4*np.pi,10)
y=np.exp(-x)*np.sin(x)
plt.figure()
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(0,4*np.pi,10)
y=np.exp(-x)*np.sin(x)
z=np.exp(-x)
plt.figure()
plt.plot(x,y)
plt.plot(x,z,'r--',label='$ e^(x)$')
plt.legend(['y','z'])
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x=np.random.randint(250,600,30, dtype=np.int)         #independent
y=np.random.normal(x * 0.8 + 10, 35).astype(np.int)   #dependent

plt.scatter(x,y)
plt.title("x vs y")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x=np.sort(np.linspace(0,2*np.pi,50))
ydata=np.random.normal(np.sin(x),0.10)

x_uji=np.linspace(0,2*np.pi,100)
y_uji=np.sin(x_uji)

plt.scatter(x,ydata,label="data")
plt.plot(x_uji,y_uji,color='red',label="uji")
plt.legend(loc="best")
plt.xlabel("x")
plt.ylabel("ydata")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()

linreg.fit(x.reshape(-1,1),ydata)
yreg=linreg.predict(x_uji.reshape(-1,1))

plt.scatter(x,ydata,label="train")
plt.plot(x_uji,y_uji,color='red',label="uji")
plt.plot(x_uji,yreg,color='black',label="regresi")
plt.legend(loc="best")
plt.xlabel("x")
plt.ylabel("ydata")
plt.show()

print("Koefesien b1:", linreg.coef_)
print("Intercept:", linreg.intercept_)
```

```python
from sklearn.preprocessing import PolynomialFeatures
transformer = PolynomialFeatures(degree=3)

x_polinomial = transformer.fit_transform(x.reshape(-1,1))
x_uji_polinomial = transformer.fit_transform(x_uji.reshape(-1,1))

linreg.fit(x_polinomial,ydata)
yreg=linreg.predict(x_uji_polinomial)

plt.scatter(x,ydata,label="data")
plt.plot(x_uji,y_uji,color='red',label="uji")
plt.plot(x_uji,yreg,color='black',label="polinom derajat 3")
plt.legend(loc="best")
plt.xlabel("x")
plt.ylabel("ydata")
plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
df = pd.DataFrame([[151,63],[174,81],[158,56],[176,91],[138,47],[156,57],[179,76],[163,72],[152,62],[151,48]])
df.columns = ['x', 'y']
x_train = df['x'].values[:,np.newaxis]
y_train = df['y'].values
lm = LinearRegression()
lm.fit(x_train,y_train) #fase training
print("Coefficient :"  + str(lm.coef_))
print("Intercept : " + str(lm.intercept_))
x_test = [[170],[171]] #data yang akan diprediksi
p = lm.predict(x_test) #fase prediksi
print(p) #hasil prediksi
#prepare plot
pb = lm.predict(x_train)
dfc = pd.DataFrame({'x': df['x'],'y':pb})
plt.scatter(df['x'],df['y'])
plt.plot(dfc['x'],dfc['y'],color='red',linewidth=2)
plt.xlabel('Tinggi dalam cm')
plt.ylabel('Berat dalam Kg')
plt.show()
```
